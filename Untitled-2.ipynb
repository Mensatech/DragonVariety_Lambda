```python
import json
```

```python

def extract_keys_from_json(file_path, keys):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
        
        # Iterate over each object in the array and extract keys
        extracted_data = [{key: obj.get(key, None) for key in keys} for obj in data]
        return extracted_data

    except FileNotFoundError:
        return "File not found."
    except json.JSONDecodeError:
        return "Invalid JSON file."
```

```python
keys_to_extract = [
    "amountPaid", "balance", "billingAddress", "contactName",
    "dueDate", "email", "inventoryStatus", "InvoiceDate",
    "isCancelled", "isCompleted", "isInvoiced", "isPrioritized",
    "lines", "orderDate", "orderFreight", "orderNumber",
    "orderRemarks", "paymentStatus", "phone", "requestedShipDate",
    "shippingAddress", "subTotal", "tax1", "tax1Name", "total","timestamp"
]
```

```python
extracted_data = extract_keys_from_json('listSalesOrder.json', keys_to_extract)
```
